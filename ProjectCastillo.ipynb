# Linear Regression

```python
dfp = df[['winner_rank_points', 'loser_rank_points']]
dfp = dfp.dropna()
X1 = dfp.iloc[:, :-1].values  
y1 = dfp.iloc[:, 1].values
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=0)  
regressor = LinearRegression()  
regressor.fit(X1_train, y1_train) 
print(regressor.intercept_) 
print(regressor.coef_) 
```

```python
y1_pred = regressor.predict(X1_test)  
dfp = pd.DataFrame({'Actual': y1_test, 'Predicted': y1_pred})  
dfp
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y1_test, y1_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y1_test, y1_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y1_test, y1_pred))) 
```

```python
dfp = df[['w_svpt', 'l_svpt']]
dfp = dfp.dropna()
X2 = dfp.iloc[:, :-1].values  
y2 = dfp.iloc[:, 1].values
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=0)  
regressor = LinearRegression()  
regressor.fit(X2_train, y2_train) 
print(regressor.intercept_) 
print(regressor.coef_) 
```

```python
y2_pred = regressor.predict(X2_test)  
dfp = pd.DataFrame({'Actual': y2_test, 'Predicted': y2_pred})  
dfp
#test value is the actual values of loser service point
#pred value is the predicted value of loser service point
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y2_test, y2_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y2_test, y2_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y2_test, y2_pred))) 
```

# Multiple Regression

```python
# Here, we are predicting winner rank points based on winner ace, service point and 1st In
df4= df[['w_ace', 'w_svpt', 'w_1stIn','winner_rank_points',]]
df4= df4.dropna()
X4 = df4[['w_ace','w_1stIn','w_svpt']]
y4 = df4['winner_rank_points']
X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y4, test_size=0.2, random_state=0)
regressor = LinearRegression()  
regressor.fit(X4_train, y4_train)
coeff_df4 = pd.DataFrame(regressor.coef_, X4.columns, columns=['Coefficient'])
#to see the coefficients the regression model 
#has chosen
coeff_df4  
```

```python
y4_pred = regressor.predict(X4_test)  
dfp2 = pd.DataFrame({'Actual': y4_test, 'Predicted': y4_pred})  
dfp2
#Data shows that the correlation of the training variables to the test variable is yucky
```
